```python
import librosa
import numpy as np
import deepspeech

# Load the pre-trained DeepSpeech model
model_file_path = r'C:\Users\Omar Muhammed\Downloads\u\deepspeech-0.9.3-models.pbmm'
model = deepspeech.Model(model_file_path)

# If you have a scorer file
scorer_file_path = r'C:\Users\Omar Muhammed\Downloads\u\deepspeech-0.9.3-models.scorer'
model.enableExternalScorer(scorer_file_path)

audio_file_path = 'extracted_audio.wav'

# Load the audio file at its original sample rate
audio, sr_orig = librosa.load(audio_file_path, sr=None, mono=True)  # Ensure the audio is mono

# Check if the audio is not silent
if np.max(audio) == 0:
    print("The audio is silent.")
else:
    # Resample the audio to match the model's sample rate
    audio_resampled = librosa.resample(audio, orig_sr=sr_orig, target_sr=model.sampleRate())
    
    # Normalize the audio to have a consistent volume level
    audio_resampled = librosa.util.normalize(audio_resampled)

    # Define the chunk size in seconds and calculate the chunk size in frames
    chunk_size_seconds = 60
    chunk_size_frames = int(chunk_size_seconds * model.sampleRate())

    # Calculate the number of chunks needed
    num_chunks = int(np.ceil(len(audio_resampled) / chunk_size_frames))

    # Initialize an empty string for the transcription
    full_text = ''

    for i in range(num_chunks):
        # Extract the chunk
        start = i * chunk_size_frames
        end = start + chunk_size_frames
        audio_chunk = audio_resampled[start:end]

        # Convert the chunk to 16-bit integers
        audio_chunk_int16 = np.int16(audio_chunk * 32767)

        # Transcribe the audio chunk using DeepSpeech
        chunk_text = model.stt(audio_chunk_int16)
        full_text += chunk_text + ' '

    # Output the full transcription
    if full_text:
        print("Transcribed text:", full_text)
    else:
        print("Transcription is empty. The speech might not be recognized.")
```

```python
import spacy

# Load an English model
nlp = spacy.load('en_core_web_sm')

# Process the text
doc = nlp(full_text)

# Extract named entities, for example
for entity in doc.ents:
    print(f"{entity.text} ({entity.label_})")
```

```python
from pydub import AudioSegment

def convert_mp3_to_wav(mp3_file_path, wav_file_path):
    # Load the MP3 file
    audio = AudioSegment.from_mp3(mp3_file_path)
    # Export as WAV
    audio.export(wav_file_path, format="wav")

# Usage
mp3_file_path = r'C:\Users\Omar Muhammed\output_audio.mp3'
wav_file_path = r'C:\Users\Omar Muhammed\output_audio.wav'
convert_mp3_to_wav(mp3_file_path, wav_file_path)
```

```python
def segment_audio(audio_path, min_silence_len=700, silence_thresh=-40):
    audio = AudioSegment.from_wav(audio_path)
    return split_on_silence(audio, min_silence_len=min_silence_len, silence_thresh=silence_thresh)

segments = segment_audio(wav_file_path, min_silence_len=700, silence_thresh=-40)

# For example, print the length of the first segment
if segments:  # Check if any segments were found
    print(f"First segment duration: {len(segments[0])}ms")
```
